```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x=np.arange(0,5,0.1)
x
```

```python
y=x**3
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,"ro")
plt.show()
```

```python
plt.plot(x,y,color="black",marker="o")
plt.show()
```

```python
plt.plot(x,y,color="black",linewidth=5,label="x^3")
plt.ylabel("X^3")
plt.xlabel("X")
plt.title("Matplotlib demo")
y2=x**2
plt.plot(x,y2,"r",label="x^2")

plt.legend()
plt.grid()
plt.text(2,20,"text",fontsize=15)
plt.show()
```

```python
#pie graphs
sizes=[3,4,6,2]
colors=["blue","purple","red","aqua"]
plt.pie(sizes)
plt.show()
```

```python
sizes=[3,4,6,2]
colors=["pink","purple","red","aqua"]
plt.pie(sizes,colors=colors)
plt.show()
```

```python
sizes=[3,4,6,2]
colors=["pink","purple","red","aqua"]
plt.title("my plot")
plt.axis("equal")
plt.pie(sizes,colors=colors)
plt.show()
```

```python
#add labels to classes
sizes=[3,4,6,2]
colors=["pink","purple","red","aqua"]
plt.title("my plot")
plt.axis("equal")
#add labels to classes
labels=["A","B","C","D"]
plt.pie(sizes,colors=colors,labels=labels)
plt.show()
```

```python
#add % to classes
sizes=[3,4,6,2]
colors=["pink","purple","red","aqua"]
plt.title("my plot")
plt.axis("equal")
#add labels to classes
labels=["A","B","C","D"]

plt.pie(sizes,colors=colors,labels=labels,autopct="%.2f%%")
plt.show()
```

```python
#add explode a class
sizes=[3,4,6,2]
colors=["pink","purple","red","aqua"]
plt.title("my plot")
plt.axis("equal")
#add labels to classes
labels=["A","B","C","D"]

explode=[0.2,0.1,0,0]
plt.pie(sizes,colors=colors,labels=labels,autopct="%.2f%%",explode=explode)
plt.show()
```

```python
#mking classes in clockwise direction
sizes=[3,4,6,2]
colors=["pink","purple","red","aqua"]
plt.title("my plot")
plt.axis("equal")
#add labels to classes
labels=["A","B","C","D"]

explode=[0.2,0.1,0,0]
plt.pie(sizes,colors=colors,labels=labels,autopct="%.2f%%",explode=explode,counterclock=False)
plt.show()
```
